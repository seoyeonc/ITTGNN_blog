## 제주에너지공사 풍력 발전량

[출처](https://www.jejuenergy.or.kr/index.php/contents/infomation/openingpublicdata/monthlypower?act=view&seq=2719&bd_bcid=monthlypower)

```python
dfs = []
for i in range(1, 5):
    file_path = f'./data_example/2022년 {i}분기 풍력 발전량.csv'
    
    df = pd.read_csv(file_path)
    
    dfs.append(df)

combined_df = pd.concat(dfs, ignore_index=True)
```

```python
(combined_df==0).sum()
```

```python
plt.plot(combined_df['신창풍력'])
```
